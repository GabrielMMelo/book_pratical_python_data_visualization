```python
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
```

```python
#pip install matplotlib
```

### Preparing data

```python
x = np.arange(5)
```

```python
print(x)
print(type(x))
```

```python
y=x
```

```python
plt.plot(x, y, 'o--')
plt.plot(x, x**2, 's--')
plt.plot(x, 2**x, 'd--')
plt.show()
```

#### Save it to a file

```python
plt.plot(x, y, 'o--', x, x**2, 's--', x, 2**x, 'd--')  # same as above
plt.savefig('out/first_matplotlib_chart.png')
plt.show()
```

### Grid, Axes and Labels

#### Grid

```python
plt.plot(x, y, 'o--')
plt.grid(True)  # activate grid
plt.show()
```

#### Axis

```python
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.grid(True)
plt.axis([0, 4, 0, 64])  # set axis with [xmin, xmax, ymin, ymax]
print(plt.axis())
plt.show()
```

##### or

```python
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.grid(True)
plt.xlim([0, 4])
plt.ylim([0, 64])
print(plt.axis())
plt.show()
```

#### Labels for axis

```python
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.grid(True)
plt.xlabel('x = np.arange(5)')
plt.xlim([0, 4])
plt.ylabel('y = f(x)')
plt.ylim([0, 64])
plt.title('Simple Plot Demo')
plt.show()
```

#### Labels for lines

```python
plt.plot(x, x**2, label='x**2')  # set label for each curve
plt.plot(x, x**3, label='x**3')
plt.plot(x, 2*x, label='x*2')
plt.plot(x, 2**x, label='2**x')
plt.legend()                     # enable labels for curves
plt.grid(True)
plt.xlabel('x = np.arange(5)')
plt.xlim([0, 4])
plt.ylabel('y = f(x)')
plt.ylim([0, 64])
plt.title('Simple Plot Demo')
plt.show()
```

##### or

```python
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.legend(['x**2', 'x**3', '2*x', '2**x'], loc='upper left')  # pass legends as array. `loc` is optional
plt.grid(True)
plt.xlabel('x = np.arange(5)')
plt.xlim([0, 4])
plt.ylabel('y = f(x)')
plt.ylim([0, 64])
plt.title('Simple Plot Demo')
plt.show()
```

### Colors, Styles, and Markers

#### Colors

```python
x = np.arange(5)
y = x
plt.plot(x, y+1, 'g')
plt.plot(x, y+0.5, 'y')
plt.plot(x, y, 'r')
plt.plot(x, y-0.2, 'c')
plt.plot(x, y-0.4, 'k')
plt.plot(x, y-0.6, 'm')
plt.plot(x, y-0.8, 'w')
plt.plot(x, y-1, 'b')
plt.show()
```

##### or

```python
plt.plot(x, y+1, 'g', x, y+0.5, 'y', x, y, 'r', x, y-0.2, 'c',
x, y-0.4, 'k', x, y-0.6, 'm', x, y-0.8, 'w', x, y-1, 'b')
plt.show()
```

#### Line Styles

```python
plt.plot(x, y, '-', x, y+1, '--', x, y+2, '-.', x, y+3, ':')
plt.show()
```

#### Markers

```python
plt.plot(x, y, '.')
plt.plot(x, y+0.5, ',')
plt.plot(x, y+1, 'o')
plt.plot(x, y+2, '<')
plt.plot(x, y+3, '>')
plt.plot(x, y+4, 'v')
plt.plot(x, y+5, '^')
plt.plot(x, y+6, '1')
plt.plot(x, y+7, '2')
plt.plot(x, y+8, '3')
plt.plot(x, y+9, '4')
plt.plot(x, y+10, 's')
plt.plot(x, y+11, 'p')
plt.plot(x, y+12, '*')
plt.plot(x, y+13, 'h')
plt.plot(x, y+14, 'H')
plt.plot(x, y+15, '+')
plt.plot(x, y+16, 'D')
plt.plot(x, y+17, 'd')
plt.plot(x, y+18, '|')
plt.plot(x, y+19, '_')
plt.show()
```

#### Combining all

```python
plt.plot(x, y, 'mo--')
plt.plot(x, y+1 , 'g*-.')
plt.show()
```

##### Or using parameters

```python
plt.plot(x, y, color='g', linestyle='--', linewidth=1.5,
 marker='^', markerfacecolor='b', markeredgecolor='k',
 markeredgewidth=1.5, markersize=5)
plt.grid(True)
plt.show()
```

#### Ticks

```python
x = y = np.arange(10)
plt.plot(x, y, 'o--')
plt.xticks(range(len(x)), ['a', 'b', 'c', 'd', 'e', 'f', 'g',
'h', 'i', 'j'])
plt.yticks(range(0, 10, 1))
plt.show()
```
