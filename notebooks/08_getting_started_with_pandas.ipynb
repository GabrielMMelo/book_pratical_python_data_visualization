```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

#### Simple `Series`

```python
ts = pd.Series(np.random.randn(5))
```

```python
ts
```

```python
ts.cumsum()
```

```python
ts.cumsum().plot(marker='o', linestyle='--')
plt.grid(True)
plt.show()
```

#### Simple `DataFrame`

```python
df1 = pd.DataFrame(np.random.randn(10, 4),  # negative values
 columns=['A', 'B', 'C', 'D'])

df2 = pd.DataFrame(np.random.rand(10, 4),
 columns=['A', 'B', 'C', 'D'])
```

```python
df1
```

#### Bar chart

```python
df1.plot.bar()  # or df1.plot(kind='bar')
plt.title("Bar chart", loc='center')
plt.show()
```

#### Horizontal bar chart

```python
df1.plot(kind='barh')
plt.title("Horizontal bar chart", loc='center')
plt.show()
```

#### Stacked bar

```python
df2.plot(kind='bar', stacked=True)
plt.show()
```

#### Horizontal stacked bar

```python
df2.plot(kind='barh', stacked=True)
plt.show()
```

#### Area chart

```python
df2.plot(kind='area')
plt.show()
```

#### Unstacked area chart

```python
df2.plot(kind='area', stacked=False)  # by default, it is stacked
plt.show()
```

#### Pie chart

```python
df2.transpose()[0].plot(kind='pie')
plt.show()
```
