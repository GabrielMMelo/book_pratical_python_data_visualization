#### Strange colors

```python
img2 = plt.imread('../img/pirate.tif')
plt.imshow(img2)
plt.axis(False)
plt.show()
```

```python
img2 = plt.imread('../img/pirate.tif')
plt.imshow(img2, cmap=plt.cm.gray)
plt.axis(False)
plt.show()
```

```python
plt.subplots_adjust(hspace=0.4, wspace=0.1)
plt.subplot(1, 3, 1)
plt.title('Original')
plt.axis(False)
plt.imshow(img2)
plt.subplot(1, 3, 2)
plt.title('Fixed cmap')
plt.axis(False)
plt.imshow(img2, cmap=plt.cm.gray)
plt.subplot(1, 3, 3)
plt.title('Testing other cmaps')
plt.axis(False)
plt.imshow(img2, cmap='Reds')
plt.show()
```
