### Working with Graphs and Networks

```python
#%pip install networkx
```

```python
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
G = nx.Graph()
```

```python
print(type(G))
print(G.nodes())
print(G.edges())
print(type(G.nodes()))
print(type(G.edges()))
```

#### Add nodes

```python
## one by time
G.add_node('a')
```

```python
## multiple nodes at once
G.add_nodes_from(['b', 'c'])
```

#### Add edges

```python
## one by time
G.add_edge(1, 2)  # if nodes dont exist, they will be created
edge = ('d', 'e')
G.add_edge(*edge)
edge = ('a', 'b')
G.add_edge(*edge)
```

```python
## multiple edges at once
G.add_edges_from([('a', 'c'), ('c', 'd'),
 ('a', 1), (1, 'd'),
 ('a', 2)])
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
```

#### Relabeling nodes

```python
G = nx.path_graph(4)
cities = {0: 'Mumbai', 1: 'Hyderabad',
 2: 'Banglore', 3: 'New York'}
H=nx.relabel_nodes(G, cities)
```

```python
nx.draw(H, with_labels=True)
plt.show()
```
